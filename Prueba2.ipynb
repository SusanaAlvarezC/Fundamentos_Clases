### **2.3. Calcule la masa muscular por fila con una función como**

$$I_{m.m.} = \frac{Weight}{{Height}^2}$$

```python
def imm(row, variable, variable2):

  """
    Calcula el índice de masa muscular (IMM) utilizando los valores de
    las columnas 'variable' y 'variable2' en la fila.

    Args:
        row (pandas.Series): Fila del DataFrame con los valores a
            utilizar para el cálculo.
        variable (str): Nombre de la columna que contiene el valor
            del numerador (peso).
        variable2 (str): Nombre de la columna que contiene el valor
            del denominador al cuadrado (altura).

    Returns:
        float or None: El valor calculado del índice de masa muscular
            o None si el denominador es cero.

    """
  if(row.loc[variable2]>0):
    return row.loc[variable]/row.loc[variable2]**2


"""
Aplicar la función imm a cada fila del DataFrame 'athletes' utilizando
apply con axis=1. Los argumentos 'weight' y 'height' se pasan a la
función como argumentos. El resultado se agrega como una nueva columna
llamada 'Imm' en el DataFrame.
"""
athletes['Imm'] = athletes.apply(imm, axis = 1,
                                 args =['weight', 'height'])

athletes
```

```python
import numpy as np
import pandas as pd

# Con la función pd.read se extraen los datos que están generados en el
# archivo nba
nba = pd.read_csv(
    'https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv'
    ,index_col=0)

# Con index_col se salta la primera columna que no tiene informacion
# relevante
nba.head()
```

```python
# El método.dropna() eliminará todas las filas que contengan
# al menos un valor faltante (NaN) en cualquier columna.
nba_filtered = nba.dropna()

nba_filtered.head()
```

### **3.1. Equipos con salarios mas altos**
Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)

```python
df_highest_salary = nba_filtered.groupby(['Team'])[['Salary']].sum()

df_highest_salary_sorted = df_highest_salary.sort_values(
    by='Salary', ascending=False)

# Cambiamos el nombre de la columna 'id' a 'cantidad'
df_highest_salary_sorted = df_highest_salary_sorted.rename(
    columns={'Team': 'Nombre del Equipo', 'Salary': 'Cantidad Dinero'})

# Solicitar al usuario el número de filas que desea visualizar
top_n = int(input('Escriba el top de salarios que desea visualizar:'))

# Mostrar las primeras n filas según la entrada del usuario
df_highest_salary_sorted.head(top_n)
```

### **3.2. Separe la variable Age en tres grupos**



#### 3.2.1. Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información

```python
nba_filtered['Age-Group'] = pd.cut(
    nba_filtered['Age'], bins=3, labels=[
        'Primer Grupo', 'Segundo Grupo', 'Tercer grupo'])

nba_filtered
```

#### 3.2.2. Presentamos la edad mínima y máxima que se asignó a cada grupo

### **3.3. Tabla dinamica salario por equipo y posición**
Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
# Primero calculamos el promedio del salario por equipo y posicion,
# con sort_Values ordenamos por salario descendente y con head tomamos
# los primeros 5 valores
avg_salary = nba_filtered.groupby(['Team', 'Position'])[['Salary']]\
.mean().sort_values(by = 'Salary',ascending = False).head(5)

avg_salary
```

Creamos tabla dinamica con el resultado anterior

```python
pivot_table = pd.pivot_table(avg_salary, values='Salary',
                             index='Team', columns='Position',
                             fill_value=0)

pivot_table
```
